```python
#importing other libraries
import pandas as pd
import numpy as np
```

```python
#reading data from the dataframe stored int toronto.csv
df = pd.read_csv('toronto.csv')
df.head()
```

```python
#reading data of latitudes and longitudes from the given path in a new dataframe
path='http://cocl.us/Geospatial_data'
dfgeo=pd.read_csv(path)
dfgeo.head()
```

```python
#renaming Postal code in dfgeo dataframe to PostalCode
dfgeo.rename(columns={'Postal Code': 'PostalCode'}, inplace=True)
```

```python
#merging the dataframes df and dfgeo into on new dataframe
newdf=pd.merge(df,dfgeo,on='PostalCode',how='left')
newdf.head(10)
```

```python
#saving contents of the new dataframe "newdf" to toronto_part2.csv file
newdf.to_csv('toronto_part2.csv',index=False)
```
